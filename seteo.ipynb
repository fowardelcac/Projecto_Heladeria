```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('ventas.xlsx')
df.head()
```

```python
fecha = []
tabla = pd.DataFrame()

column_n = len(df.columns)

for j in range(len(df) - 1):
    tabla[df.loc[j+1][0]] = df.loc[j+1][1:column_n]

for i in range(len(df.columns) - 1):
    f = df.loc[0][i+1]
    fecha.append(f)

tabla['Fecha'] = fecha
tabla.set_index('Fecha', inplace = True)
tabla.fillna(0, inplace = True)
```

```python
tabla.head()
```

```python
pd.options.display.float_format = '{:.3f}'.format
```

```python
tabla.describe()
```

```python
desv = tabla.describe().std()
sab = pd.DataFrame()
sab['Std'] = desv.loc[50 < desv]
sab
```

```python
plt.figure(figsize=(8, 6))
plt.title("Distribucion del ANANA")
plt.hist(tabla['ANANA'])
```

```python
col = tabla.columns

data = tabla.copy()
data = np.where(data > 1000, data / 1000, np.where(data > 50, data / 100, data))
data = pd.DataFrame(data, columns=[col])
data.index = tabla.index
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.hist(tabla['ANANA'])
ax1.set_title('Anana sin correccion')

# Graficar en el segundo eje
ax2.hist(data['ANANA'])
ax2.set_title('Anana corregido')

# Ajustar el layout
plt.tight_layout()
```
