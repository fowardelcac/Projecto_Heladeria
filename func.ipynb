```python
def agrupar_mensual(data):
    df = data.copy()
    #df.set_index('Fecha')
    return df.groupby(pd.Grouper(key='Fecha', freq='M')).sum()
    
def venta_anual(df, helado):
    return df[helado].sum()

def venta_sort(data, fecha):
    df = data.set_index('Fecha')
    df = df.loc[fecha]
    return df.sort_values(ascending = False)
```
